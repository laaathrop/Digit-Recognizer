```python
# Import
import pandas as pd
```

```python
# Read in the data
df = pd.read_csv('./data/train.csv')
```

```python
df.head()
```

```python
# Get pixel values for a single label
df[df['label'] == 9].loc[11, :]
```

```python
# Counts of numbers labels 1-9. The classes seem to be approximately balanced.
df['label'].value_counts()
```

```python
# Get value counts for a few pixels
print(df['pixel0'].value_counts())
print()
print(df['pixel38'].value_counts())
print()
print(df['pixel358'].value_counts())
```

Each row in the dataset consists of a label (the numerical digit to be predicted) and 784 pixel columns. For any given label (single row) the cell for each pixel will contain a value between 0 and 255 that corresponds to the shade of the pixel.

```python
# Get value counts for each column of dataset
[df[column].value_counts() for column in df]
```
